```python
import psycopg2
import pandas as pd
```

```python
def create_database():
    #Conectar a base de dados padrão
    conn = psycopg2.connect('host=127.0.0.1 dbname=postgres user=postgres password=root')
    conn.set_session(autocommit=True)
    cur = conn.cursor()
    # Criar database
    cur.execute('DROP DATABASE accounts')
    cur.execute('CREATE DATABASE accounts')
    # Fechar a conexão com o banco de dados padrão
    conn.close()
    # Conectar ao banco de dados criado
    conn = psycopg2.connect('host=127.0.0.1 dbname=accounts user=postgres password=root')
    cur = conn.cursor()
    return cur, conn
```
